```python
df = pd.read_csv('hmeq.csv')
df.head()
```

### Handling Missing Values ##

```python
df.isna().sum()/len(df)
```

```python
df2 = df.dropna()
```

```python
df.shape,df2.shape
```

```python
df2.describe()
```

### Taking Care OF Categorical Values ###

```python
df2['JOB'] = df2.JOB.map({'Self':1,'Other':2,'Sales':3,'Office':4,'Mgr':5,'ProfExe':6})
df2['REASON'] = df2.REASON.map({'HomeImp':2,'DebtCon':1})
df2.head()
```

### Count Plot for Number of Defaulters and Payers

```python
%matplotlib inline
sns.countplot(x='BAD',data=df2,hue='BAD')
```

### Visualisng Distribution and Relationship 

```python
sns.scatterplot(x='LOAN',y='MORTDUE',hue='BAD',data=df2)
```

### Splitting Dataset in Train-Test-Validation

```python
from sklearn.model_selection import train_test_split as tts
X = df2.iloc[:,1:]
y= df2.iloc[:,0]
xtrain,xtest ,ytrain,ytest = tts(X,y)
```

### Applying Algorithms###

##### 1) SVM
#####  2) Decision Tree
#####  3) Random Forest
#####  4) Xgboost
#####  5) Isolation Forest

### <------------------------------------------>
### Support Vector Machines ###

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(xtrain,ytrain)
ypred=svm.predict(xtest)
```

#### Accuracy and Matrix for SVM

```python
from sklearn.metrics import confusion_matrix as cm
print(cm(ytest,ypred))
svm.score(xtest,ytest)
```

### Saving Classifier ###

```python
pd.to_pickle(svm,'hmeq_svm_90.pickle')
```

## Repeating Steps for Every Algorithms ##

```python
import xgboost
from xgboost import XGBClassifier as XG
xg = XG()
xg.fit(xtrain,ytrain)
ypred_xg=xg.predict(xtest)
print(cm(ytest,ypred_xg))
xg.score(xtest,ytest)
```

```python
pd.to_pickle(xg,'XG_hmeq_93.pickle')
```


### Decision Trees ###

```python
from sklearn.tree import DecisionTreeRegressor as DT
dt = DT()
dt.fit(xtrain,ytrain)
y_pred_dt = dt.predict(xtest)
print(cm(ytest,y_pred_dt))
dt.score(xtest,ytest)
```

```python
from sklearn.ensemble import GradientBoostingClassifier as GBR
gb = GBR()
gb.fit(xtrain,ytrain)
y_pred_gb = gb.predict(xtest)
print(cm(ytest,y_pred_gb))
gb.score(xtest,ytest)
```

```python
pd.to_pickle(gb,'hmeq_gb_94.pickle')
```
